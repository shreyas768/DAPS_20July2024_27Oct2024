## Multiple Conditions


* AND -->> &
* OR -->> |
* NOT -->> ~

```python
df[(df['smoker'] == 'Yes') & (df['time']=='Dinner')]
```

```python
#all the male customer with total bill being more than 40

df[(df['total_bill']>40)&(df['sex']=='Male')]
```

```python
df['sex'].value_counts()
```

```python
df[(df['total_bill']>40)&(df['sex']!='Male')]
```

```python
df[(df['total_bill']>40) | ~(df['sex']=='Male')]
```

```python
#df for weekend
df[(df['day'] == 'Sat') | (df['day'] == 'Sun') ]
```

# isin()

```python
weekend = ['Sat','Sun']

df[df['day'].isin(weekend)]
```
