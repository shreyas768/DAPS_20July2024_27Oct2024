```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

```python
#basic matplotlib commands

plt.plot(x,y)
plt.xlabel('X axis title here')
plt.ylabel('Y axis title here')
plt.title('Set title here')
plt.xlim(0,6)
plt.ylim(0,10)
plt.savefig('example.png')
plt.show()
```

### Object Oriented Method in Matplotlib

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
```

```python
#create blank canvas

fig = plt.figure()

#add set of axes to figure
axes = fig.add_axes([0,0,1,1]) #left, bottom, width, height

#plot on this set of axes
axes.plot(x,y)

plt.show()
```

```python
#create blank canvas

fig = plt.figure()

#add set of axes to figure
axes1 = fig.add_axes([0,0,1,1]) #left, bottom, width, height
axes2 = fig.add_axes([1.2,1.2,0.5,0.5])

#plot on this set of axes
axes1.plot(a,b)
axes2.plot(x,y)

#Big figure
axes1.set_xlabel('A label')
axes1.set_ylabel('B label')
axes1.set_title('Big Fig')

#Small figure
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Small Fig')
axes2.set_xlim(4,8)
axes2.set_ylim(5,15)

plt.show()
```
