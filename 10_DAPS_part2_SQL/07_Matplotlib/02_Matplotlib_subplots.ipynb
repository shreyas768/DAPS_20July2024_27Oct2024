```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a **4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'red')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('demo figure')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols=2, figsize = (12,8))

axes[0][0].plot(a,b)
axes[1][1].plot(b,a)
axes[1][0].plot(x,y)
axes[0][1].plot(y,x)

axes[1][0].set_title('this is row 1 and col 0')
axes[1][0].set_xlabel('X')
axes[1][0].set_ylabel('Y')

fig.suptitle('Super title',fontsize = 20)
fig.supxlabel('Super X Label')
fig.supylabel('Super Y Label')



plt.tight_layout()
plt.show()
```
