```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a **4
x = np.arange(0,10)
y = 2*x
```

```python
#Legends

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(a,a**2, label='a squared')
ax.plot(a,a**3, label = 'a cubed')
ax.legend(loc = (0.9,0.5))
```

```python
fig, ax = plt.subplots()

ax.plot(a,b,color='tab:blue', marker='o')
ax.plot(x,y,'r.-')
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color = 'black', alpha= 0.82 , lw = 3.90, ls= '--')
ax.plot(x, x-1, color = '#B54848', alpha= 0.82 , lw = 5, ls = '-.', marker = 's', ms = 50, mec = 'blue', mfc = 'red', markeredgewidth = 8)
```
