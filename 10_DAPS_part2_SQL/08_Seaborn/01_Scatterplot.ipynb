```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.info()
```

```python
#scatterplot
plt.figure(figsize=(14,9))
sns.scatterplot(x = 'salary',y='sales',data=df, size = 'training level', palette='Set1', hue = 'work experience', style = 'division')
plt.savefig('example_scatter.png')
```
