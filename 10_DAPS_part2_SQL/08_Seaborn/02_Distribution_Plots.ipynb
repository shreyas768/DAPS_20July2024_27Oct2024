```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dm_office_sales.csv')
df.head()
```

## Rugplot

```python
sns.rugplot(x='salary',data=df, height = 0.5)
```

### Displot and Histogram

```python
plt.figure(figsize=(14,9))
sns.displot(data=df,x='salary', kde = True)
```

```python
sns.histplot(data=df,x='salary',bins=100, kde=True)
```

```python
plt.figure(figsize=(14,9))
sns.set_theme(style="whitegrid", palette="Set1")
sns.histplot(data=df,x='salary',bins=100)
```

```python
plt.figure(figsize=(14,9))
sns.set_theme(style="whitegrid", palette="pastel")
sns.histplot(data=df,x='salary',bins=20, color='red', edgecolor='black', lw = 8, ls = '--')
```

### KDE Plot

```python
np.random.seed(50)

ages = np.random.randint(0,100,200)
ages
```

```python
ages_df = pd.DataFrame(ages, columns = ['age'])
```

```python
sns.rugplot(data=ages_df,x='age')
```

```python
sns.histplot(data=ages_df,x='age',kde=True)
```

```python
sns.kdeplot(data=ages_df,x='age', clip = [0,90])
```

```python
sns.kdeplot(data=ages_df,x='age', bw_adjust=0.1)
```

```python
sns.kdeplot(data=ages_df,x='age', bw_adjust=1, shade = True)
```

```python
#2-D KDE plots

sns.kdeplot(data=df, x='salary',y='sales')
```
