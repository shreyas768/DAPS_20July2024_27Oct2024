```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dm_office_sales.csv')
df.info()
```

```python
#countplot

plt.figure(figsize=(10,3))
sns.countplot(x='level of education',data=df)
```

```python
plt.figure(figsize=(10,3))
sns.countplot(x='level of education',data=df, hue='division')
```

```python
#barplot

fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(x='level of education',data=df,y='salary', estimator=np.mean, ax = ax)
ax.bar_label(ax.containers[-1], fmt='Mean:\n%.2f', label_type='center')
plt.xticks(rotation = 90)
plt.show()
```

```python
#barplot

fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(x='level of education',data=df,y='salary', hue = 'division', estimator=np.mean, ax = ax, errorbar = 'sd')
ax.bar_label(ax.containers[-1], fmt='Mean:\n%.2f', label_type='center')
plt.xticks(rotation = 90)
plt.show()
```
