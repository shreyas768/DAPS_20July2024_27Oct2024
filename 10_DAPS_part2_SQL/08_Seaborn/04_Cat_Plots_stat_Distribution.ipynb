```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dm_office_sales.csv')
df.info()
```

```python
#boxplot

sns.boxplot(data=df,x = 'division', y = 'sales', hue = 'level of education', width = 0.5)
```

```python
sns.boxplot(data=df,y = 'division', x = 'sales', hue = 'level of education',orient='h')
```

```python
#violin plot

sns.violinplot(data=df,x = 'division', y = 'sales',hue = 'level of education')
```

```python
df1 = pd.read_csv('StudentsPerformance.csv')
df1.head()
```

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,hue='race/ethnicity')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,hue='gender', split=True)
```

```python
sns.violinplot(x='parental level of education',y='math score',data=df1,hue='gender', split=True, inner='quartile')
```

```python
sns.violinplot(y='parental level of education',x='math score',data=df1,hue='gender', split=True, inner='quartile', bw = 0.1)
```

```python
#swarmplots
sns.swarmplot(x = 'math score',y = 'race/ethnicity',data=df1,size=2.5, hue='gender',dodge=True)
```
