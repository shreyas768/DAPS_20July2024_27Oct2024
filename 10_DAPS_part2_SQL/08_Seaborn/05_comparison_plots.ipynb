```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('StudentsPerformance.csv')
df1.head()
```

```python
#jointplot()
sns.jointplot(data=df1,x='math score',y='writing score')
```

```python
sns.jointplot(data=df1,x='math score',y='writing score',kind='hex')
```

```python
sns.jointplot(data=df1,x='math score',y='writing score',kind='kde')
```

```python
sns.jointplot(data=df1,x='math score',y='writing score',kind='reg')
```

```python
#pairplot

sns.pairplot(df1,hue='gender')
```

```python
sns.pairplot(df1,hue='gender',corner=True,diag_kind='hist')
```

```python
#facetgrid

g = sns.FacetGrid(data=df1,col='gender',row='lunch')
g = g.map(plt.scatter,'math score','reading score')
g.add_legend()
```
