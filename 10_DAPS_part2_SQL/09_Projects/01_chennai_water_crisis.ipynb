
## Importing the relevant Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('chennai_reservoir_levels.csv')
df.head()
```

```python
df.info()
```

```python
type(df['Date'][0])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df.info()
```

```python
df['Year'] = df['Date'].apply(lambda x:x.year)
```

```python
df.head()
```

```python
df['Month'] = df['Date'].apply(lambda x:x.month)
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
df['Total'] = df[['POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM']].sum(axis=1)
df.head()
```

```python
df['Year'].unique()
```

```python
yearwise = df.groupby('Year',as_index=False).mean()
```

```python
yearwise
```

```python
plt.figure(figsize=(14,8))
sns.barplot(x='Year',y='Total',data=yearwise)
plt.title('Mean Yearly water level in all four reservoir')
plt.show()
```

```python

yearwise.plot(x='Year',y=['POONDI', 'CHOLAVARAM', 'REDHILLS', 'CHEMBARAMBAKKAM'], figsize=(14,8))
```

```python
yearwise
```

```python
dec = df[df['Month']==12]
```

```python
plt.figure(figsize=(14,8))
sns.barplot(x='Year',y='Total',data=dec)
plt.title('Mean water level in Feb in all four reservoir from 2004 to 2019')
plt.show()
```

```python
monthwise = df.groupby('Month',as_index=False).mean()
plt.figure(figsize=(14,8))
sns.barplot(x='Month',y='Total',data=monthwise)
plt.title('Mean Monthly water level in all four reservoir')
plt.show()
```

## Inferences :

1. Although chennai receives some rainfall in July, Aug, Sep due to SW monsoon, majority of rainfall happens in Oct to Dec due to NE monsoon
2. From 2006 to 2010 monsoon was significant but in last few years, decline in water level is in line with declining monsoon - both NE and SW
