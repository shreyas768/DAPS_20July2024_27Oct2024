```python
df = pd.merge(fandango,all_sites,on='FILM',how='inner')
df.head()
```

## Normalize columns to ratings 0-5

```python
df['RT_Norm'] = np.round(df['RottenTomatoes']/20,1)
```

```python
df['RTU_Norm'] = np.round(df['RottenTomatoes_User']/20,1)
```

```python
df['Meta_Norm'] = np.round(df['Metacritic']/20,1)
```

```python
df['Meta_U_Norm'] = np.round(df['Metacritic_User']/2,1)
```

```python
df['IMDB_Norm'] = np.round(df['IMDB']/2,1)
```

```python
df.head()
```

```python
df.columns
```

```python
norm_scores = df[['STARS', 'RATING','RT_Norm', 'RTU_Norm', 'Meta_Norm', 'Meta_U_Norm', 'IMDB_Norm']]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(data=norm_scores, x='STARS',label='STARS',fill=True)
sns.kdeplot(data=norm_scores, x='RATING',label='TRUE RATING',fill=True)
sns.kdeplot(data=norm_scores, x='RT_Norm',label='RT_Norm',fill=True)
sns.kdeplot(data=norm_scores, x='RTU_Norm',label='RTU_Norm',fill=True)
sns.kdeplot(data=norm_scores, x='Meta_Norm',label='Meta_Norm',fill=True)
sns.kdeplot(data=norm_scores, x='Meta_U_Norm',label='Meta_U_Norm',fill=True)
sns.kdeplot(data=norm_scores, x='IMDB_Norm',label='IMDB_Norm',fill=True)
plt.title('Fandango vs All')
plt.legend()
plt.show()
```
